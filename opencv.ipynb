```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_bgr=cv2.imread("E:/DATA/lena.jpg")
plt.imshow(img_bgr)
plt.show()
```

```python
color=('b','g', 'r')
```

```python
height, width, _ = img_bgr.shape
for i in range(0, height - 1):
    for j in range(0, width - 1):
             pixel = img_bgr[i, j]
             pixel[0] = 255 - pixel[0]
    pixel[1] = 255 - pixel[1]
    pixel[2] = 255 - pixel[2]
    img_bgr[i, j] = pixel
```

```python
plt.imshow(img_bgr)
plt.show()
```

```python
color = ('b', 'g', 'r')
  
for i, col in enumerate(color):
      
    histr = cv2.calcHist([img_bgr], 
                         [i], None,
                         [256],
                         [0, 256])
      
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
    plt.show()
```
